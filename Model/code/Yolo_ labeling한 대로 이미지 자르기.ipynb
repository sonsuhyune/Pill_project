```python
import cv2
import os
from matplotlib import pyplot as plt
from xml.etree.ElementTree import parse
from xml.etree.ElementTree import Element, dump
import xml.etree.ElementTree as ET


filename_list=os.listdir("train_data/pill_result/xml")
filename_list.sort()
for file_name in filename_list:
    img_filename=file_name.split(".")[0]
    if len(img_filename) < 2:
        continue
    img = cv2.imread("train_data/pill_box/gray_img/"+img_filename+".jpg"); ###### 
    
    doc = ET.parse("train_data/pill_box/xml/"+file_name)
    root = doc.getroot()
    object_tag = root.findall("object") #pill
    
    obj=object_tag[0]
    box_tag_pill = obj.find("bndbox")
    xmin_pill = box_tag_pill.findtext("xmin")
    ymin_pill = box_tag_pill.findtext("ymin")
    xmax_pill = box_tag_pill.findtext("xmax")
    ymax_pill = box_tag_pill.findtext("ymax")
    print("-"*50)
    print(xmin_pill, ymin_pill, xmax_pill,ymax_pill) 
    print("Save!: "+"train_data/pill_result/img/"+img_filename+".jpg")
    try: #.jpg,.JPG  error  try, except 
        cv2.imwrite("train_data/pill_result/img/"+img_filename+".jpg",img[int(ymin_pill):int(ymax_pill),int(xmin_pill):int(xmax_pill)])
    except:
        img = cv2.imread("train_data/pill_box/gray_img/"+img_filename+".JPG"); ###### 
        cv2.imwrite("train_data/pill_result/img/"+img_filename+".JPG",img[int(ymin_pill):int(ymax_pill),int(xmin_pill):int(xmax_pill)])
    
    
    #print("Save!: "+"train_data/pill_result/img/"+file_name)
```
