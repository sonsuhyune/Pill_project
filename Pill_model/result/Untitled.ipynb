```python
f = open("check_all_with_right.txt", 'r')
lines = f.readlines()
pred_list = []
gt_list = []
for line in lines:
    if len(line.split(' ')) == 1:
        pass
    
    
    elif line.split(' ')[0] == 'pred:':
        pred_num = len(line.split(' '))
        temp = []
        if pred_num == 2:
            temp.append(line.split(' ')[1][2:-3])
            #print(line.split(' ')[1][2:-3])
        elif pred_num == 3:
            temp.append(line.split(' ')[1][2:-2])
            temp.append(line.split(' ')[2][1:-3])
            
            #print(line.split(' ')[1][2:-2])
            #print(line.split(' ')[2][1:-3])
            
        elif pred_num == 4:
            temp.append(line.split(' ')[1][2:-2])
            temp.append(line.split(' ')[2][1:-2])
            temp.append(line.split(' ')[3][1:-3])
           
            
        else:
            print("---pred_num---: ",pred_num)
            print(line)
        pred_list.append(temp)
            
    elif line.split(' ')[0] == 'gt:':
        gt_num = len(line.split(' '))
        temp=[]
        if gt_num == 2:
            temp.append(line.split(' ')[1][2:-3])
        elif gt_num == 3:
            temp.append(line.split(' ')[1][2:-2])
            temp.append(line.split(' ')[2][1:-3])  
        elif gt_num == 4:
            temp.append(line.split(' ')[1][2:-2])
            temp.append(line.split(' ')[2][1:-2])
            temp.append(line.split(' ')[3][1:-3])    
        else:
            print("---gt_num---: ",gt_num)
            print(line)
        gt_list.append(temp)
```

```python
f = open("custom_source.txt", 'w')
ft = open("custom_target.txt", 'w')
for pred, gt in zip(pred_list, gt_list):
    if(len(pred) != len(gt)):
        continue
    else:
        for p, g in zip(pred,gt):
            if(len(p)==len(g)):
                print(p,g)
                f.write(p+"\n")
                ft.write(g+'\n')
```
