```python
lst = [] #정제
for front, back, typ, shape in zip(df_small['표시앞'], df_small['표시뒤'], df_small['성상'], df_small['의약품제형']):
    if "정제" in typ:
        lst.append((front, back, typ, shape))
    elif "캡슐" in typ:
        pass
    elif "캅셀제" in typ:
        pass
    elif "정" in typ:
        lst.append((front, back, typ, shape))
```

```python
lst_no_mark = []
for front, back, shape in lst:
    if front == '-' and back == '-':
        lst_no_mark.append((front, back, shape))
    if front == '마크' and back == '-':
        lst_no_mark.append((front, back, shape))
    if front == '-' and back == '마크':
        lst_no_mark.append((front, back, shape))
```
